```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Customers=pd.read_csv("D:\case study one for students\Customer.csv")
print(Customers.head())
```

```python
Product_Hierarchy=pd.read_csv("D:\case study one for students\prod_cat_info.csv")
print(Product_Hierarchy.head())
```

```python
Transactions=pd.read_csv("D:\case study one for students\Transactions.csv")
print(Transactions.head())
```

Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure 
tokeep all customers who have done transactions with us and select the join type accordingly.

```python
customer_transaction=pd.merge(Customers,Transactions,left_on='customer_Id',right_on='cust_id')
print(customer_transaction)
```

```python
Customer_Final=pd.merge(customer_transaction,Product_Hierarchy,left_on=['prod_cat_code','prod_subcat_code'],right_on=['prod_cat_code','prod_sub_cat_code'],how='left')
# pd.set_option('display.max_columns',None)
print(Customer_Final)
```

 Generate histograms for all continuous variables and frequency bars for categorical variables.

```python
# Bar Chart for Categorical Variables
cat_cols=['Gender','Store_type','prod_cat','prod_subcat','city_code']
for col in cat_cols:
    plt.figure()
    Customer_Final[col].value_counts().plot(kind='bar')
    print(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout() # Labels automatically adjust ho jayenge
    plt.show()
    
```

```python
    
# Histogram for Continuous Variables
cont_cols = ['Qty', 'Rate', 'Tax', 'total_amt']
for col in cont_cols:
    plt.figure()
    Customer_Final[col].hist(color='y')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

Which City code has the maximum customers and what was the percentage of customers from that city?

```python
city_customer_counts=Customer_Final.groupby(['city_code'])['customer_Id'].nunique() # -> count the number of unique customers
print(city_customer_counts)

max_city=city_customer_counts.idxmax()
print("Maximum Customers in this city:",max_city)

city_percent=(city_customer_counts/city_customer_counts.sum())*100
print(city_percent)

max_percent=city_percent.max()
print("Percentage of customers:",max_percent)
```

Which store type sells the maximum products by value and by quantity?

```python
sales_by_value=Customer_Final.groupby(['Store_type'])['total_amt'].sum()
pd.options.display.float_format='{:.2f}'.format
print(sales_by_value)
print(sales_by_value.idxmax(),sales_by_value.max())
sales_by_count=Customer_Final.groupby(['Store_type'])['Qty'].sum()
print(sales_by_count)
print(f"{sales_by_count.idxmax()} sells the maximum products by value and quantity")

```

How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts? 

```python
filtered_data=Customer_Final.loc[Customer_Final['total_amt']>0]
print(filtered_data)
```

```python
customer_counts=filtered_data.groupby(['customer_Id'])['transaction_id'].nunique()
print(customer_counts)
```

```python
result=customer_counts[customer_counts>10]
print(result)
print("Number of customers have more than 10 Unique transactions:",result.size)
```
